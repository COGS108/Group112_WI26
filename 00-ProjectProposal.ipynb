## Research Question

How do regional and seasonal differences in national parks in the US from 2000 to 2024 affect both recreation visitation and hours spent per visit?

## Background and Prior Work

As a team the famous U.S National Parks grabbed our attention during the process of selecting a topic. We believed that there were many interesting datasets and analysis that could be made regarding this topic. The U.S National Parks are spread among different states. However, they are different in many aspects. For instance, some national parks are more populated and easily accessible, while some are in less populated areas and are harder to get to. Each region's national park has its own unique features. Generally speaking, National Parks that are located near the major cities will be more accessible than the ones that are located in the more remote areas, which the latter will require greater effort and planning to visit. 
In addition to the regional difference, National Parks also have seasonal differences. Each national park has its own peak and off-season based on what tourists plan to do there. These factors greatly influence tourists in deciding whether they choose to visit which national park and the duration of their stay there. Hence, this project aims to explore how regional and seasonal differences in the US National Parks from 2000 to 2024 affect recreational visitation and hours spent per visit. 


https://www.nps.gov/subjects/socialscience/visitor-use-statistics-dashboard.htm 

A report by the National Park Service (NPS) shows that the reported number of recreational visits is 331.9 million in the year 2024, with visitors spending approximately 1.4 billion hours in national parks. Additionally, visitors distributed their visits throughout the year, where more than half of the parks show peak periods in February to June, and October to December. This shows that the peak season is during spring and fall, which is the opposite to popular belief that the peak season is summertime. Instead, tourists visit throughout the year in different periods, and over half of the park shows that spring and fall had an above-average number of visits. 

https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0128226

The National Park Service also found out that climate change not only affect resources of national parks, but they discovered that it influenced visitation patterns. Their data shows that 95% of parks has great relationship between air temperature and visitation. Visitation declines as the monthly average temperature is over 80F. This shows that a lot of the park visitors avoid hot weather, which leads to this phenomenon.
