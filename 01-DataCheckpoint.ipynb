## Background and Prior Work

As a team, the famous U.S National Parks grabbed our attention during the process of selecting a topic. We believed that there were many interesting datasets and analysis that could be made regarding this topic. The U.S National Parks are spread among different states. However, they are different in many aspects. For instance, some national parks are more populated and easily accessible, while some are in less populated areas and are harder to get to. Each region's national park has its own unique features. Generally speaking, National Parks that are located near the major cities will be more accessible than the ones that are located in the more remote areas, which the latter will require greater effort and planning to visit. 

In addition to the regional difference, National Parks also have seasonal differences. Each national park has its own peak and off-season based on what tourists plan to do there. These factors greatly influence tourists in deciding whether they choose to visit which national park and the duration of their stay there. Hence, this project aims to explore how regional and seasonal differences in the US National Parks from 2000 to 2024 affect recreational visitation and hours spent per visit. 

A report by the National Park Service (NPS) shows that the reported number of recreational visits is 331.9 million in the year 2024, with visitors spending approximately 1.4 billion hours in national parks.<a name="ref1"></a>[1](#footnote1) Additionally, visitors distributed their visits throughout the year, where more than half of the parks show peak periods in February to June, and October to December. This shows that the peak season is during spring and fall which is the opposite to popular belief that the peak season is summertime. Instead, tourists visit throughout the year in different periods, and over half of the park shows that spring and fall had an above-average number of visits.

The National Park Service also found out that climate change not only affects resources of national parks, but they discovered that it influenced visitation patterns. Their data shows that 95% of parks has great relationship between air temperature and visitation.<a name="ref2"></a>[2](#footnote2) Visitation declines as the monthly average temperature is over 80F. This shows that a lot of the park visitors avoid hot weather, which leads to this phenomenon.

A report from the NPS in 2021 shows that the most visited national park is the Great Smoky Mountain National Park with 14 million recreation visits, with Zion National Park in second with 5 million recreation visits.<a name="ref3"></a>[3](#footnote3) The main reason why the Great Smoky Mountain National Park is so popular is due to its geographical location. It is located within a day's drive for most US population including cities like Chicago, Atlanta, Nashville, etc. Which made it very easy and accessible for a wide range of the population to visit without having to drive too far.<a name="ref4"></a>[4](#footnote4)

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## importing pandas for the data
import pandas as pd

## loading in the dataset
national_parks = pd.read_csv('data/00-raw/2000-2024 National Park Data.csv')

## Grabbing and renaming the columns that we are going to be analyzing
national_parks_cleaned = national_parks[['ParkName', 'Region', 'State', 'Year', 'Month', 'RecreationVisits', 'RecreationHours']]
national_parks_cleaned = national_parks_cleaned.rename(columns={
    "ParkName": "Park Name",
    "RecreationVisits": "Recreation Visits",
    "RecreationHours": "Recreation Hours"
})

## Changing the Visists # and Hours # into ints for calculation purpose
national_parks_cleaned["Recreation Visits"] = pd.to_numeric(
    national_parks_cleaned["Recreation Visits"].astype(str).str.replace(",", "")
)

national_parks_cleaned["Recreation Hours"] = pd.to_numeric(
    national_parks_cleaned["Recreation Hours"].astype(str).str.replace(",", ""),
)

## Adding the Hours Per Visit Column
national_parks_cleaned['Hours Per Visit'] = (national_parks_cleaned['Recreation Hours'] / national_parks_cleaned['Recreation Visits'])

## Loading the Data into the processed folder
processed_path = "data/02-processed/nps_visitation_2000_2024_processed.csv"
national_parks_cleaned.to_csv(processed_path, index = False)
```
