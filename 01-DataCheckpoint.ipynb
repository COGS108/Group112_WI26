- Dataset #1
  - Dataset Name: National Park Service Visitor Use Statistics (IRMA)
  - Link to the dataset: https://irma.nps.gov/Stats/SSRSReports/National%20Reports/Query%20Builder%20for%20Public%20Use%20Statistics%20(1979%20-%20Last%20Calendar%20Year)
  - Number of observations:
    Monthy obersvations, for each park unit, for this project we will take data from (2000 -2024)
  - Number of variables:
     - National Parks
     - Year
     - Month
     - Recreation visits
     - Non-recreation visits
     - Recreation visitor hours
     - Non-recreation visitor hours
     - Overnight stays (by type) 
  - Recreation visits: are defiend as the number of people entering a park for recreational purposes in a given month. This will be used as the primary measure of park visitation
  - Visitor hours: used to construct a new variable. Average time spent per visit = recreation vistior hours/recreation visits. This will provides a proxy for visit length since the dataset does not directly report average duration
  - Year & Month: is to analyze trends over time, create season varaibles
     - Winter:Dec-Feb
     - Spring:Mar-May
     - Summer:Jun-Aug
     - Fall:Sep-Nov
  - Park Name used as a unique identifier to merge with external dadtasets (region, park type, weather)
  - Overnight stays:can be used additional measure of park use and to distinguish between day use and overnight visitation patterns
  - Non-recreation visits: will not be used in the main visitation measure because they include commuting and administrative use rather than tourism-related visits
  - Shortcomings is vistiations does not mean unique visitations, the dataset only takes into account of recreation visits not individual people. We cannot measure number of unique visitors and repeat visitation behavior.
  - There is no direction "average length of stay" because we only have recreation vistiro hours and recreation visits so we must construct average stay = vistiors hours/visits. This assumes that all visitor hours come form recreation visitors and reporitng is accruate accross parks
  - There is no visor characterisitcs, can not anaylze differences by age, income, residency (local vs toruosits), group size, transportation mode. So the analysis is limited to park patterns not people.
  - There is no weather in the dataset, you cannot directly measure temperature, rain, snow and extreme events, so you cannot test weather effects unless you merge naother dataset
  - Potential differences in counting methods across parks. They may use different tools like traffic counters, fee recipts, or manual estimates which can create inconsistency in visitation estimates especially for older data
  - Some missing or suppressed data: for certain praks or early years: some months may be missing, some overnight categories may be blank.
  - Measures park-level use, not specific activites: Based on the dataset we won't be able to tell from hiking vs camping vs sightseeing, or which areas insidee the park are crowded
